```python
import pandas as pd

data = {
    "name": ["Amit", "Neha", "Raj", "Sara", "Kunal"],
    "city": ["Pune", "Mumbai", "Pune", "Delhi", "Pune"],
    "age": [22, 25, 23, 26, 24],
    "salary": [30000, 45000, 35000, 50000, 40000]
}

df = pd.DataFrame(data)

df
```
