```python
import numpy as np

prices = [55000, 55000, 70000, 52000, 32003, 55000, 120000, 30003]

arr = np.array(prices)

print(arr)

# condition applied to entire array → creates boolean mask
print(arr > 60000)

# arr[mask] → filters array and returns only masked elments which are true or satisfy the condition
print(arr[ (arr > 40000) & (arr <80000) ])
```

```python
#How many orders are above 60000?
print(np.size(arr[arr > 60000]))
```

```python
#Any price below 40000 should be treated as 40000 (minimum selling price policy)
arr[arr<40000] = 40000
print(arr)
```

```python
# Create labels for each price in arr:
# if price ≥ 60000 → "Premium"
# otherwise → "Normal"

labels = np.where(arr>=60000, "Premium", "Normal")
print(labels)
```

```python
#2d NumPy array
data = np.array([
    [55000, 2],
    [70000, 5],
    [30000, 1]
])

print(data)
```
